```python
df["km"] = df["km"].str.split("km").str[0]
```

```python
df["km"] = df["km"].str.replace(",","")
df
```

```python
df.drop("hp", axis=1 , inplace=True)
df
```

```python
df["kw"] = df["kw"].str.split("kW").str[0]
df
```

```python
df["Previous Owners"] = df["Previous Owners"].str.lstrip("\n")
df["Previous Owners"] = df["Previous Owners"].str.rstrip("\n")
df
```

```python
df["Previous Owners"].value_counts(dropna=False)
```

```python
def orta(i) :
    try :
        return i[1]
    except :
        return i
```

```python
df["Body Color"] = df["Body Color"].apply(orta)
df["Body"] = df["Body"].apply(orta)
df["Gearing Type"] = df["Gearing Type"].apply(orta)
df["Fuel"] = df["Fuel"].apply(orta)
df
```

```python
df.loc[df["Fuel"].str.contains("Diesel"), "Fuel"] = "Diesel"
df.loc[df["Fuel"].str.contains("CNG|Others|Biogas|Electric|Domestic|LPG"), "Fuel"] = "Others"
df.loc[df["Fuel"].str.contains("Benzine|Super|Gasoline"), "Fuel"] = "Benzine"
df
```

```python
def ortadaki(k):
    try:
        return k[0]
    except :
        return k
```

```python
df["Displacement"] = df["Displacement"].apply(ortadaki)
df["Displacement"] = df["Displacement"].str.strip("\n")
df
```

```python
def ortada(l):
    try:
        return l[0]
    except :
        return l
```

```python
df["Cylinders"] = df["Cylinders"].apply(ortada)
df["Cylinders"] = df["Cylinders"].str.strip("\n")
df["Drive chain"] = df["Drive chain"].apply(ortada)
df["Drive chain"] = df["Drive chain"].str.strip("\n")
df["Nr. of Doors"] = df["Nr. of Doors"].apply(ortada)
df["Nr. of Doors"] = df["Nr. of Doors"].str.strip("\n")
df["Gears"] = df["Gears"].apply(ortada)
df["Gears"] = df["Gears"].str.strip("\n")
df
```
