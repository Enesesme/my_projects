```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Sampiyonlar Ligi 55-16.csv")
df
```

```python
df.head()
```

```python
final = df[["Year","Round","Team 1","Team 2","FT","ET","P"]]
```

```python
final.head()
```

```python
final.info()
```

```python
final.sort_values("Year",inplace = True)
```

```python
final
```

```python
final["Round"].unique()
```

```python
final = final[final["Round"].str.contains("Final")]
final
```

```python
final["Round"].unique()
```

```python
final["Team 1"] = final["Team 1"].str.split(" › ").str[0]
```

```python
final["Team 2"] = final["Team 2"].str.split(" › ").str[0]
```

```python
final
```

```python
final["FT"] = final["FT"].str.strip(" (*)")
```

```python
final.head()
```

```python
final["ET"].fillna("0-0",inplace = True)
```

```python
final
```

```python
final["ET"] = final["ET"].str.strip(" (a.e.t.)")
```

```python
final
```

```python
final["P"].fillna("0-0",inplace = True)
```

```python
final
```

```python
final["P"] = final["P"].str.strip(" (pen.)")
```

```python
final
```

```python
final.reset_index(drop = True,inplace = True)
final
```

```python
final["Team1FT"] = final["FT"].str.split("-").str[0]
final["Team2FT"] = final["FT"].str.split("-").str[1]
final["Team1ET"] = final["ET"].str.split("-").str[0]
final["Team2ET"] = final["ET"].str.split("-").str[1]
final["Team1P"] = final["P"].str.split("-").str[0]
final["Team2P"] = final["P"].str.split("-").str[1]
final
```

```python
final[final["Round"].str.contains("Replay")]
```

```python
final[final["Year"] == 1974]
```

```python
final["ET"][19] = final["FT"][18]
```

```python
final[final["Year"] == 1974]
```

```python
final.drop(index = 18,inplace = True)
```

```python
final.reset_index(drop = True,inplace = True)
```

```python
final.head()
```

```python
final[["Team1FT","Team2FT","Team1ET","Team2ET","Team1P","Team2P"]] = final[["Team1FT","Team2FT","Team1ET","Team2ET","Team1P","Team2P"]].astype("int64")
```

```python
final.info()
```

```python
final["Team1Total"] = final["Team1FT"] + final["Team1ET"] + final["Team1P"]
final["Team2Total"] = final["Team2FT"] + final["Team2ET"] + final["Team2P"]
final["Averaj"] = final["Team1Total"] - final["Team2Total"]
final
```

```python
final.head()
```

```python
final = final[["Year","Team 1","Team 2","FT","ET","P","Averaj"]]
final
```

```python
final.loc[final['Averaj'] > 0, "Sampion"] = final["Team 1"]
final.loc[final['Averaj'] < 0, "Sampion"] = final["Team 2"]
```
